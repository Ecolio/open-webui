```python
import ollama

# Define the questions
questions = [
    "Are there any repeated sections or sentences in the document?",
    "Are there any contradictions or logical errors in the document?",
    "Does the narrative flow smoothly and logically throughout the document?",
    "Are there any parts of the document that are unclear or difficult to understand?"
]

# Define the language models to use
models = ['mistral', 'phi3', 'qwen2']

# Initialize an empty list to store all answers from each model for each question
all_answers = []

# Iterate over each question
for question in questions:
    print(f"Question: {question}")  # Print the original text of the question
    question_answers = []  # Reset the list of answers for this question
    # For each language model, get the answer for this question
    for model in models:
        print(f"Model: {model}")  # Print the name of the current model
        messages = [
            {
                'role': 'system',
                'content': f'Use the following document to answer the question: {text}',
            },
            {
                'role': 'user',
                'content': question,
            },
        ]
        # print(messages)
        response = ollama.chat(model=model, messages=messages)
        print(response)
        # Store the answer in the list of answers for this question
        question_answers.append(response['message']['content'])
    # Synthesize the answers from all models for this question using Llama3
    synthesis_question = "Summarize the following feedback on the document: " + ", ".join(question_answers)
    messages = [
        {
            'role': 'system',
            'content': f'Use the following feedback to answer the question: {", ".join(question_answers)}',
        },
        {
            'role': 'user',
            'content': synthesis_question,
        },
    ]
    print("Model: llama3")  # Print the name of Llama3 model for this part
    #print(messages)
    final_response = ollama.chat(model='llama3', messages=messages)
    print(final_response)
    # Store the synthesized answer in the list of all answers
    all_answers.append(final_response['message']['content'])

# Define a final question to summarize all the feedback from Llama3's synthesis
final_question = "Summarize the overall feedback on the document: " + ", ".join(all_answers)

# Get the final summary from the model Llama3-gradient
messages = [
    {
        'role': 'system',
        'content': f'Use the following feedback to answer the question: {", ".join(all_answers)}',
    },
    {
        'role': 'user',
        'content': final_question,
    },
]
print("Model: llama3-gradient")  # Print the name of Llama3-gradient model for this part
#print(messages)
final_synthesis = ollama.chat(model='llama3-gradient', messages=messages)
print(final_synthesis['message']['content'])
```

```python
import ollama

# Define the questions
questions = [
    "Are there any repeated sections or sentences in the document?",
    "Are there any contradictions or logical errors in the document?",
    "Does the narrative flow smoothly and logically throughout the document?",
    "Are there any parts of the document that are unclear or difficult to understand?"
]

# Define the language models to use
models = ['mistral', 'phi3', 'qwen2']

# Initialize an empty list to store all answers from each model for all questions
all_answers = []

# Iterate over each model
for model in models:
    print(f"Model: {model}")  # Print the name of the current model
    model_answers = []
    # For each question, get the answer from this model
    for question in questions:
        print(f"Question: {question}")  # Print the original text of the question
        messages = [
            {
                'role': 'system',
                'content': f'Use the following document to answer the question: {text}',
            },
            {
                'role': 'user',
                'content': question,
            },
        ]
        response = ollama.chat(model=model, messages=messages)
        model_answers.append(response['message']['content'])
    # After answering all questions with this model, add these answers to the list of all answers
    all_answers.append(model_answers)

# Now that we have answered all questions with each model, synthesize the answers with Llama3
gemma2_synthesis = []
for i in range(len(questions)):  # For each question
    question_answers = [model_answers[i] for model_answers in all_answers]  # Get the answers to this question from all models
    synthesis_question = "Summarize the following feedback on the document: " + ", ".join(question_answers)
    messages = [
        {
            'role': 'system',
            'content': f'Use the following feedback to answer the question: {", ".join(question_answers)}',
        },
        {
            'role': 'user',
            'content': synthesis_question,
        },
    ]
    print("Model: gemma2")  # Print the name of Llama3 model for this part
    final_response = ollama.chat(model='gemma2', messages=messages)
    gemma2_synthesis.append(final_response['message']['content'])

# Finally, use Llama3-gradient to summarize all the feedback from Llama3's synthesis
final_question = "Summarize the overall feedback on the document: " + ", ".join(gemma2_synthesis)
messages = [
    {
        'role': 'system',
        'content': f'Use the following feedback to answer the question: {", ".join(gemma2_synthesis)}',
    },
    {
        'role': 'user',
        'content': final_question,
    },
]
print("Model: llama3-gradient")  # Print the name of Llama3-gradient model for this part
final_synthesis = ollama.chat(model='llama3-gradient', messages=messages)
print(final_synthesis['message']['content'])
```
