```python

from langchain_community.document_loaders import PyPDFLoader

# Load the PDF and extract text
loader = PyPDFLoader("/Users/lijou/Documents/Documents/Courses/Coursera/AdvancedDjango/Inro_Document.pdf")
pages = loader.load()
pdf_text = '\n'.join([page.page_content for page in pages])
```

```python
import ollama


question = 'is there contradictions in the document?'

messages = [
  {
    'role': 'system',
    'content': f'Use the following document to answer the question: {pdf_text}',
  },
  {
    'role': 'user',
    'content': question,
  },
]

response = ollama.chat(model='llama3-gradient', messages=messages)
print(response['message']['content'])
```

```python
import ollama

# Define the questions
questions = [
    "Are there any repeated sections or sentences in the document?",
    "Are there any contradictions or logical errors in the document?",
    "Does the narrative flow smoothly and logically throughout the document?",
    "Are there any parts of the document that are unclear or difficult to understand?"
]

# Initialize an empty list to store the answers
answers = []

# Iterate over each question
for question in questions:
    messages = [
        {
            'role': 'system',
            'content': f'Use the following document to answer the question: {pdf_text}',
        },
        {
            'role': 'user',
            'content': question,
        },
    ]
    print(messages)
    response = ollama.chat(model='mistral', messages=messages)
    # Store the answer in the answers list
    answers.append(response['message']['content'])

# Define a final question to summarize all the feedback
final_question = "Summarize the following feedback on the document: " + ", ".join(answers)

# Get the final summary from the model
messages = [
    {
        'role': 'system',
        'content': f'Use the following feedback to answer the question: {", ".join(answers)}',
    },
    {
        'role': 'user',
        'content': final_question,
    },
]
print(messages)
final_response = ollama.chat(model='llama3-gradient', messages=messages)
print(final_response['message']['content'])
```
